```python
# importiere Softwarepakete
from sentence_transformers import SentenceTransformer, util

# fixe Variablen deklarieren
model_name = 'distiluse-base-multilingual-cased-v1'

# Instantiiere das NN-Modell
model = SentenceTransformer(model_name)
```

Ein paar Satzbelege mit dem Lemma "Regenschirm" aus https://www.dwds.de/r/?q=Regenschirm&corpus=kernbasis

```python
# examples = [
#     'Nebel nimmt Scharfschützen die Sicht, Regen prasselt so originalgetreu auf die Soldaten nieder, dass man sich unweigerlich einen Regenschirm herbeisehnt, dazu Donnergrollen und in der Ferne stampft ein Zug über eine Stahlbrücke.',
#     'Der voranstürmende Große ist der Theoretiker des Freundeskreises, August Wilhelm Schlegel, der Kleine mit dem Regenschirm Schleiermacher und der Reiter des gestiefelten Katers Ludwig Tieck.',
#     'Als der Citroën-Generaldirektor 1935 die legendäre »Ente« in Auftrag gab, schwebten ihm »vier Räder unter einem Regenschirm« vor.',
#     'Sinnbildlich für das kollektive Glück gibt es keine Regenschirme mehr, sondern »ein fortlaufendes, wasserdichtes Dach war über das ganze Trottoir niedergelassen worden und verwandelte dieses in einen gut erleuchteten und vollkommen trockenen Korridor«.',
#     'Regenschirm meets Nähmaschine auf dem Seziertisch',
#     'Das Blech fetzt, und die Streicher sind wohl gesetzt von Helmut Brandt, einst ein bedeutender Saxophonist des deutschen Nachkriegs-Jazz, der unter den Regenschirm des Rias-Tanzorchesters flüchtete.'
# ]
examples = [
    "Die Feiernden sind betrunken.",
    "Die Feiernden sind blau.",
    "Die Feiernden sind vom blauen Himmel begeistert.",
    "Der blaue Himmel ist schön."
]
```

Berechne die Cosine Similarity (S. https://en.wikipedia.org/wiki/Cosine_similarity#Definition ).

- Die 6 Satzbelege werden in 6 Einbettungsvektoren umgewandelt, die jeweils 512 Features haben. 
- Danach werden immer 2 Vektoren ausgewählt (insg. `6*(6-1)/2` Kombinationen) und jeweils die Cosine Similarity Funktionen berechnet.


```python
# Berechne die Outputvektoren
embeddings = model.encode(examples, convert_to_tensor=True)
print(f"N_examples={embeddings.shape[0]} Vektoren mit jeweils N_features={embeddings.shape[1]} Elementen")

# Berechne die Similarity Matrix
cosine_similarity = util.pytorch_cos_sim(embeddings, embeddings)
print(f"Similarity Matrix, shape={list(cosine_similarity.shape)}")
```
